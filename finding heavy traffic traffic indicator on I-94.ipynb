```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("traffic.csv",sep= ";")
df
```

```python
df.head(20)
```

```python
df["Slowness in traffic (%)"]=df["Slowness in traffic (%)"].str.replace("," ,".").astype(float)
```

```python
df["Slowness in traffic (%)"].describe()
```

```python
df["Slowness in traffic (%)"].plot(kind = "hist")
plt.title("distribution in slowness in traffic")
plt.xlabel("Slowness in traffic (%)")
plt.show()
```

```python
incidents =df.drop(["Hour (Coded)","Slowness in traffic (%)"],axis=1)
incidents
```

```python
incidents=incidents.sum(axis=0)
incidents
```

```python
plt.bar(incidents.index,incidents.values)
plt.xticks(rotation=90)
plt.show()
```
