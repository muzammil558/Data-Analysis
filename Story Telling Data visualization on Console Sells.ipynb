```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
df=pd.read_csv("console (1).csv")
df
```

```python
df=df[df["Sales"]>10]
```

```python
df=df.sort_values(by='Release_Year')
```

```python
df = df.drop(0, axis=0)
```

```python
colors=['yellow', 'yellow', 'greenyellow','greenyellow','greenyellow', 'greenyellow', 'limegreen','limegreen','green', 'green', 'green','green', 'darkslategrey','darkslategrey', 'darkslategrey','darkslategrey','darkslategrey','midnightblue', 'midnightblue', 'midnightblue','midnightblue','mediumblue','royalblue','royalblue']
```

```python
fig , ax= plt.subplots( figsize = (8,6))

plt.bar(df["Console_Name"],df["Sales"], color=colors,width=0.9)
plt.xticks(rotation=90)
# plt.xticks(df["Console_Name"], df["Release_Year"])
for location in ['left','right','bottom','top']:
    ax.tick_params(bottom = 0 , left = 0)
    ax.spines[location].set_visible(False)
yellow_patch = mpatches.Patch(color='yellow', label='1977-1985')
greenyellow_patch = mpatches.Patch(color='greenyellow', label='1988-1990')
limegreen_patch = mpatches.Patch(color='limegreen', label='1994-1996')
green_patch = mpatches.Patch(color='green', label='2000-2001')
darkslategrey_patch = mpatches.Patch(color='darkslategrey', label='2004-2006')
midnightblue = mpatches.Patch(color='midnightblue', label='2011-2013')
mediumblue_patch = mpatches.Patch(color='mediumblue', label='2017')
royalblue_patch = mpatches.Patch(color='royalblue', label='2020')
ax.legend(bbox_to_anchor=(.92,1),loc='upper left', borderaxespad=0.,\
                                                                      handles=[yellow_patch,\
                                                                               greenyellow_patch,\
                                                                               limegreen_patch,\
                                                                               green_patch,\
                                                                               darkslategrey_patch,\
                                                                               midnightblue,\
                                                                               mediumblue_patch,\
                                                                               royalblue_patch])
ax.get_legend().set_title("Launch Year")
ax.set_ylabel('Units Sold in Millions', fontsize=12)
ax.text(5, 160,"Most Sold \n Console \n   ever", alpha = 0.7) 
ax.text(7.2, 155,"--->", alpha = 0.7,rotation = 300) 
ax.text(-3.2,180 ,"Famous Consoles Sold in Last 45 Years",\
        fontdict = {"family": "Serif",'size':17,"weight":"bold"})
plt.grid(color='gray', linestyle='-', linewidth=.5, axis='y', alpha=0.2)
p = ax.bar(df["Console_Name"],df['Sales'],alpha=0)
ax.bar_label(p, label_type='center',rotation = 90,color="white",fmt='%.f'+"m")

plt.show()
```
